# Audio embeddings extraction example

This notebook is a very simple example of how to extract pre-trained OpenL3 embeddings for a bunch of copyright-free audio files.

## 1. Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import librosa
import librosa.display
from scipy.spatial import distance
from pathlib import Path
from IPython.display import Audio
from config import HOP_SIZE
```

```python
plt.rcParams['figure.figsize'] = [10, 6]
```

```python
from config import DRUMS_FILENAME, CELLO_FILENAME, ENGINE_FILENAME, SINE_FILENAME
```

## 2. Example: drums + bass guitar sample

```python
plt.title("Drums + bass guitar sample waveform")
y, sr = librosa.load("raw_data/381516__abbasgamez__drums.wav")
librosa.display.waveshow(y, color="blue")
Audio(y, rate=sr)
```

```python
drums_emb = np.load(DRUMS_FILENAME)
drums_emb
```

```python
drums_emb["timestamps"].shape
```

```python
drums_emb["embedding"].shape
```

```python
print(drums_emb["embedding"])
```

```python
plt.title("Drums sample")
sns.heatmap(drums_emb["embedding"].T, cmap="viridis")
```
